```python
import pandas as pd
import matplotlib.pyplot as plt

db_orcal = pd.read_csv('db_oracle_11g.csv')
container = pd.read_csv('dcos_container.csv')
docker = pd.read_csv('dcos_docker.csv')
mw_redis = pd.read_csv('mw_redis.csv')
os = pd.read_csv('os_linux.csv')
```

```python
docker
```

```python
docker_group = docker.groupby("name")
container_cpu = docker_group.get_group('container_cpu_used')
container_cpu = container_cpu.sort_values(by='timestamp')
docker_cpu = container_cpu.groupby('cmdb_id').get_group('docker_001')
docker_cpu["time"] = pd.to_datetime(docker_cpu.timestamp, unit='ms', origin='1970-01-01 08:00:00')
```

```python
docker_cpu
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[15,8])
plt.plot(docker_cpu.time, docker_cpu.value)
```
